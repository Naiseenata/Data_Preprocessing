```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Downloads/Employee.csv')
df
```

```python
df1= df.nunique()
print("Unique values in each feature:\n",df1)
```

```python
# Perform statistical analysis
statistics = df.describe()
print("Statistical analysis:\n", statistics)
```

```python
df.dtypes
```

```python
df=df.rename(columns={'Gender': 'Male'})
df
```

## Data Cleaning 

```python
df.isnull().sum()
```

```python
# Find missing values and treat them appropriately
df['Age'].replace(0, pd.NA, inplace=True)  # Replace 0 in 'age' with NaN
df
```

```python
df['Salary'].fillna(df['Salary'].mean(), inplace=True)  # Replace NaN values with mean of respective columns
df
```

```python
null_company=df["Company"].mode()[0]
df["Company"].fillna(null_company,inplace=True)
df
```

```python
null_place=df["Place"].mode()[0]
df["Place"].fillna(null_place,inplace=True)
df
```

```python
df['Age'].fillna(df['Age'].mean(), inplace=True)

df['Age']=df['Age'].astype(int)
df
```

```python
df.isnull().sum()
```

```python
# Remove duplicate rows
df.drop_duplicates(inplace=True)
df
```

## Data Analysis 

```python
df3 = df[(df['Age'] > 40) & (df['Salary'] < 5000)]
df3                            
```

```python
mean = np.mean(df['Salary'])
print("mean=",mean)
std = np.std(df['Salary'])
print("standard deviation=",std)
```

```python
df["zscore"]=(df.Salary-mean)/std
df
```

```python
df[(df.zscore<-3)|(df.zscore>3)]
```

```python
without_outlier=df[(df.zscore>-3)&(df.zscore<3)]
without_outlier
```

```python
#There is no outlier.
```

```python
plt.scatter(df3['Age'],df3['Salary'])
plt.xlabel('Age')
plt.ylabel('Salary')
plt.title('Age vs. Salary')
plt.show()
```

```python
place_counts = df['Place'].value_counts()
place_counts.plot(kind='bar',color="red")
plt.xlabel('Place')
plt.ylabel('Count')
plt.title('Number of People from Each Place')
plt.show()
```

```python
# Count the number of people from each company
company_counts = df['Company'].value_counts()
company_counts.plot(kind='bar',color='green')
plt.xlabel('Company')
plt.ylabel('Count')
plt.title('Number of People by Company')
plt.show()
```

```python
# Plot the histogram of salary to understand its distribution
df['Salary'].plot(kind='hist',bins=20,color='blue')
plt.xlabel('Salary')
plt.ylabel('Frequency')
plt.title('Salary Distribution')
plt.show()
```

```python
# Count the number of males and females in the dataset and represent it visually
gender_count=df['Male'].value_counts()
gender_count.plot(kind='bar',color='black')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.title('Gender Distribution')
plt.xticks(ticks=[0,1],labels=['Female', 'Male'])
plt.show()
```
